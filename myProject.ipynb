## Steps in the process:
1) Importing and cleaning: Whatever format the data is in, it usually takes some time and eﬀort to read the data, clean and transform it, and check that everything made it through the translation process intact. <br><br>

2) Single variable explorations: I usually start by examining one variable at a time, ﬁnding out what the variables mean, looking at distributions of the values, and choosing appropriate summary statistics. <br><br>

3) Pair-wise explorations: To identify possible relationships between variables, I look at tables and scatter plots, and compute correlations and linear ﬁts. <br><br>

4) Multivariate analysis: If there are apparent relationships between variables, I use multiple regression to add control variables and investigate more complex relationships. <br><br>

5) Estimation and hypothesis testing: When reporting statistical results, it is important to answer three questions: How big is the eﬀect? How much variability should we expect if we run the same measurement again? Is it possible that the apparent eﬀect is due to chance? <br><br>

6) Visualization: During exploration, visualization is an important tool for ﬁnding possible relationships and eﬀects. Then if an apparent eﬀect holds up to scrutiny, visualization is an eﬀective way to communicate results.

Histograms are useful because they make the most frequent values immediately apparent. But they are not the best choice for comparing two distributions. In this example, there are fewer “ﬁrst babies” than “others,” so some of the apparent diﬀerences in the histograms are due to sample sizes. In the next chapter we address this problem using probability mass functions.

#### Some Summary Statistics of all Live births:

```python
mean = live.prglngth.mean() 
var = live.prglngth.var() 
std = live.prglngth.std()

mean,var,std
```

Average weeks of pregnancy for any live birth is around 38.5 weeks and a standard deviation of 2-3 weeks is common.

As we said above, the sample of firsts babies is less than the others. Therefore, there is an apparent "Effect Size" that is observed. Let Cohen's effect size be denoted by D:

```python
mean1 = firsts.prglngth.mean()
mean2 = others.prglngth.mean()

var1 = firsts.prglngth.var()
var2 = others.prglngth.var()

n1,n2 = len(firsts.prglngth),len(others.prglngth)

pooled_var = (n1*var1 + n2*var2)/(n1+n2)

D = (mean1 - mean2)/np.sqrt(pooled_var)
```

```python
D
```

## Result:
Therefore, the difference between the prglngth mean of first babies and others is 0.029 times their standard deviation. <br>
i.e, The first babies do tend to arrive late (although the difference isn't that significant) <br>
Summary Statistic used : Cohen's Effect Size
